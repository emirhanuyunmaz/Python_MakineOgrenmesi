```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
veriler = pd.read_csv("odev_tenis.csv")
veriler
```

```python
windy = veriler.iloc[:,3:4].values
windy
```

```python
#Kategorik ->  Numeric
le = preprocessing.LabelEncoder()
windy[:,0] = le.fit_transform(veriler.iloc[:,3])
ohe = preprocessing.OneHotEncoder()

windy = ohe.fit_transform(windy).toarray()
windy
```

```python
windy[:,1]
```

```python
#Kategorik ->  Numeric
le = preprocessing.LabelEncoder()
play = veriler.iloc[:,4:].values
play[:,0] = le.fit_transform(veriler.iloc[:,4])
play
```

```python
#Kategorik ->  Numeric
le = preprocessing.LabelEncoder()
outlook = veriler.iloc[:,0:1].values
outlook[:,0] = le.fit_transform(veriler.iloc[:,0])

ohe = preprocessing.OneHotEncoder()

outlook = ohe.fit_transform(outlook).toarray()
outlook
```

```python
# ["overcast","rainy","sunny"]
#Kolonların oluşturulması işlemi
sonuc1 = pd.DataFrame(data=outlook,index=range(14),columns=["overcast","rainy","sunny"])
sonuc1
```

```python
#["windy"]
sonuc2 = pd.DataFrame(data=windy[:,:1],index=range(14),columns=["windy"])
sonuc2
```

```python
# ["temperature","humidity"]
sonuc3 = pd.DataFrame(data=veriler[["temperature","humidity"]] , index=range(14),columns=["temperature","humidity"])
sonuc3
```

```python
#["play"]
sonuc4 = pd.DataFrame(data=play[:,:1], index=range(14),columns=["play"])
sonuc4
```

```python
#Kolonların birleştirilmesi işlemi
s1 = pd.concat([sonuc1,sonuc3],axis=1)
s1
```

```python
#Kolonların birleştirilmesi
s2 = pd.concat([s1,sonuc2],axis=1)
s2
```

```python
s3 = pd.concat([s2,sonuc4],axis=1)
s3
```

```python
x_train , x_test , y_train , y_test = train_test_split(s2,sonuc4,test_size=0.33,random_state=0) #Model eğitimi
```

```python
x_train
```

```python
#Modelin Eğitim Aşaması 
regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)
y_pred
```
