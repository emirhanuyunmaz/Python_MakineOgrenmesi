```python
def preprocess_image(frame):
    # Görüntüyü modelin beklediği boyutlara (64x64) yeniden boyutlandır
    frame_resized = cv2.resize(frame, (32, 32))/255.0
    
    # Görüntüyü float32 formatına dönüştür ve normalize et
    frame_preprocessed = tf.keras.applications.resnet_v2.preprocess_input(
        tf.cast(frame_resized, tf.float32)
    )
    
    # 4 boyutlu tensör yap (batch size ekle)
    frame_preprocessed = tf.expand_dims(frame_preprocessed, axis=0)
    
    return frame_preprocessed
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

capture = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while capture.isOpened():
        ret, frame = capture.read()
        frame = cv2.flip(frame, 1)  # Aynalama
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # BGR'den RGB'ye çevirme
        detected_image = hands.process(image)  # El işaret noktalarını tespit et
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # Geri BGR formatına çevir

        if detected_image.multi_hand_landmarks:
            for hand_lms in detected_image.multi_hand_landmarks:
                h, w, c = image.shape  # Görüntünün yüksekliği, genişliği ve kanalları

                # El işaret noktalarının x ve y koordinatlarını topluyoruz
                landmark_x = []
                landmark_y = []
                for lm in hand_lms.landmark:
                    cx, cy = int(lm.x * w), int(lm.y * h)
                    landmark_x.append(cx)
                    landmark_y.append(cy)

                # Elin bulunduğu dikdörtgenin (bounding box) sınırlarını buluyoruz
                x_min, x_max = min(landmark_x), max(landmark_x)
                y_min, y_max = min(landmark_y), max(landmark_y)

                # Dikdörtgen alanı çizdiriyoruz
                cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

                # Elin bulunduğu bölgeyi kesiyoruz
                hand_roi = frame[y_min:y_max, x_min:x_max]

                # Eğer istenirse, el bölgesini bir modele göndermek için burada kullanabilirsiniz
                # hand_roi görüntüsünü burada bir modele input olarak gönderebilirsiniz.
                # Örneğin, el bölgesini kaydedip modeli çağırabilirsiniz:
                # prediction = model.predict(hand_roi)

                # Elin kesilmiş bölgesini başka bir pencerede gösterebiliriz
                if hand_roi.size > 0:  # Kesilen alan boş değilse göster
                    cv2.imshow('Hand ROI', hand_roi)
                    model_control = preprocess_data(hand_roi)
                    predictions = model.predict(model_control)
                    class_id = tf.argmax(predictions, axis=-1).numpy()
                    print("Classid:",class_id)
                # El işaret noktalarını görüntü üzerinde çizdir
                mp_drawing.draw_landmarks(image, hand_lms, mp_hands.HAND_CONNECTIONS,
                                          landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 255), thickness=4, circle_radius=2),
                                          connection_drawing_spec=mp_drawing.DrawingSpec(color=(20, 180, 90), thickness=2, circle_radius=2))

        # Görüntüyü göster
        cv2.imshow('Webcam', image)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

capture.release()
cv2.destroyAllWindows()
```
