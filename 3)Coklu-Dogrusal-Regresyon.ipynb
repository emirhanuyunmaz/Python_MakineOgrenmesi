```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
```

```python
veriler = pd.read_csv("veriler.csv")
```

```python
veriler
```

```python
boy = veriler[["boy"]]
```

```python
boy
```

```python
boyKilo = veriler[["boy","kilo"]] 
```

```python
boyKilo
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy="mean")
Yas = veriler.iloc[:,1:4].values
Yas
```

```python
imputer = imputer.fit(Yas[:,1:4])
Yas[:,1:4] = imputer.transform(Yas[:,1:4])
Yas
#Bu kısma kadar şayet varsa eksik veriler giderildi
```

```python
#Kategori -> Numeric
ulke = veriler.iloc[:,0:1].values
ulke
```

```python
le = preprocessing.LabelEncoder()
ulke[:,0] = le.fit_transform(veriler.iloc[:,0])
ulke
```

```python
ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()
ulke
```

```python
#cinsiyet => numeric
cinsiyet = veriler.iloc[:,-1:].values
le = preprocessing.LabelEncoder()
cinsiyet[:,-1] = le.fit_transform(veriler.iloc[:,-1])
cinsiyet
```

```python
#Kolonları oluşturulması işlemi
sonuc = pd.DataFrame(data=ulke,index=range(22),columns=["fr","tr","us"])
sonuc2 = pd.DataFrame(data=Yas , index=range(22),columns=["boy","kilo","yas"])
sonuc3 = pd.DataFrame(data=cinsiyet[:,:1],index=range(22),columns=["cinsiyet"])
```

```python
#Kolonların birleştirilmesi işlemi
s = pd.concat([sonuc,sonuc2],axis=1)
```

```python
s2 = pd.concat([s,sonuc3],axis=1)
s2
```

```python
x_train , x_test , y_train , y_test = train_test_split(s,sonuc3 ,test_size=0.33,random_state=0)
```

```python
y_train
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train) #Modelin Eğitim Aşaması
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
#Tüm verilere bakılarak boy verisini tahmin etme işlemi .
boy = s2.iloc[:,3:4].values
sol = s2.iloc[:,:3]
sag = s2.iloc[:,4:]
veri = pd.concat([sol,sag],axis=1)
yeni_x_train , yeni_x_test , yeni_y_train , yeni_y_test = train_test_split(veri,boy,test_size=0.33,random_state=0)
r2 = LinearRegression()
r2.fit(yeni_x_train,yeni_y_train)
yeni_y_pred = r2.predict(yeni_x_test)
yeni_y_pred
```

```python
#Verilerin Sonuç üzerindeki etkileri belirleme işlemi
X = np.append(arr=np.ones((22,1)).astype(int),values=veri,axis=1)
X_L = veri.iloc[:,[0,1,2,3,4,5]].values
X_L = np.array(X_L,dtype = float)
model = sm.OLS(boy,X_L).fit()
model.summary()
```
